```python
to_edit = []

for lic in list(all_main['license_number'].unique()):
#     print(lic)
    if list(export_main[export_main['License Number'] == lic]['Rollup ID'])[0] != list(all_main[all_main['license_number'] == lic]['rollup_id'])[0]:
        print('rollup', lic)
        to_edit.append(lic)
    elif list(export_main[export_main['License Number'] == lic]['License Description'])[0] != list(all_main[all_main['license_number'] == lic]['license_description'])[0]:
        print('description', lic)
        to_edit.append(lic)
    elif list(export_main[export_main['License Number'] == lic]['License Category'])[0] != list(all_main[all_main['license_number'] == lic]['license_category'])[0]:
        print('category', lic)
        to_edit.append(lic)
    elif list(export_main[export_main['License Number'] == lic]['License Status'])[0] != list(all_main[all_main['license_number'] == lic]['license_status'])[0]:
        print('status', lic)
        to_edit.append(lic)
    elif list(export_main[export_main['License Number'] == lic]['Legal Name'])[0] != list(all_main[all_main['license_number'] == lic]['legal_name'])[0]:
        print('name', lic)
        to_edit.append(lic)
```
