```python
import pandas as pd
import urllib
from sqlalchemy import create_engine
from os import listdir
from os.path import isfile, join
```

```python
# Just so I see the whole DataFrame
pd.options.display.max_columns = 50
pd.options.display.min_rows = 1000
pd.options.display.max_rows = 1000
```

```python
def compare(sql_main, to_edit):
    comparison = sql_main[['license_number', 'name_legal', 'name_dba']].merge(to_edit[['license_number', 'name_legal', 'name_dba']], on='license_number',
                                                                        suffixes=("_sql", "_new"))
    comparison['name_dba_sql'] = comparison['name_dba_sql'].fillna('')
    comparison['name_dba_new'] = comparison['name_dba_new'].fillna('')
    comparison['name_dba_sql'] = comparison['name_dba_sql'].replace('Data Not Available', '')
    comparison['name_dba_new'] = comparison['name_dba_new'].replace('Data Not Available', '')
    comparison['name_dba_sql'] = comparison['name_dba_sql'].replace('0', '')
    comparison['name_dba_new'] = comparison['name_dba_new'].replace('0', '')
    comparison['name_dba_sql'] = comparison['name_dba_sql'].replace('-', '')
    comparison['name_dba_new'] = comparison['name_dba_new'].replace('-', '')
    
    
    for x in range(len(comparison)):
        if str(comparison['name_legal_sql'][x]).lower() + " " + str(comparison['name_dba_sql'][x]).lower() == str(comparison['name_legal_new'][x]).lower() + " " + str(comparison['name_dba_new'][x]).lower():
            comparison = comparison.drop(index=x)
            
    
    return comparison.reset_index().drop(columns=['index'])
```
