```python
import pandas as pd
```

```python
df_1 = pd.read_excel('./purple_lotus_import/Transactions Report_Q1_2020.xlsx')
```

```python
df_2 = pd.read_excel('./purple_lotus_import/Transactions Report_Q1_2021.xlsx')
```

```python
df_3 = pd.read_excel('./purple_lotus_import/Transactions Report_Q2_2020.xlsx')
```

```python
df_4 = pd.read_excel('./purple_lotus_import/Transactions Report_Q2_2021.xlsx')
```

```python
df_5 = pd.read_excel('./purple_lotus_import/Transactions Report_Q3_2020.xlsx')
```

```python
df_6 = pd.read_excel('./purple_lotus_import/Transactions Report_Q3_2021.xlsx')
```

```python
df_7 = pd.read_excel('./purple_lotus_import/Transactions Report_Q4_2020.xlsx')
```

```python
df_all = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7], axis=0, ignore_index=True)
```

```python
df_all.to_csv('./purple_lotus_import/df_all.csv')
```

```python
df_all['year_month'] = df_all['Order Items Date Closed Date'].dt.strftime("%Y-%m")
```

```python
df_groupby = df_all.groupby('year_month').sum()
```

```python
df_rev_summary = df_groupby[['Order Items Total Net Sales', 'Order Items Total Gross Receipts', 'Order Items Total Cost Without Excise', 'Order Items Total Gross Profit']]
```
