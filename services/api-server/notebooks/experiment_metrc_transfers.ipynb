```python
import json
import pandas
```

```python
raw_dataframe = pandas.read_excel('data/royal_apothecary_metrc_transfers_2021_07_07.xlsx', header=0)
raw_dataframe.columns
```

```python
raw_dataframe.columns = [
    'id',
    'company_id',
    'license_id',
    'created_date',
    'delivery_id',
    'manifest_number',
    'transfer_payload',
    'vendor_id',
    'transfer_type',
    'created_at',
    'updated_at',
    'lab_results_status',
    'shipment_type_name',
    'shipment_transaction_type',
    'company_created_at',
    'company_updated_at',
    'company_name',
    'company_identifier',
    'package_product_category_name',
    'package_shipper_wholesale_price',
    'package_shipped_quantity',
]
raw_dataframe.columns
```

```python
dataframe = raw_dataframe[[
    'id',
    'company_id',
#     'license_id',
    'created_date',
#     'delivery_id',
    'manifest_number',
#     'transfer_payload',
#     'vendor_id',
    'transfer_type',
    'created_at',
    'updated_at',
    'lab_results_status',
    'shipment_type_name',
    'shipment_transaction_type',
#     'company_created_at',
#     'company_updated_at',
    'company_name',
    'company_identifier',
    'package_product_category_name',
    'package_shipper_wholesale_price',
    'package_shipped_quantity',
]]
dataframe
```

```python
pandas.set_option('display.max_rows', 10)
fn = lambda row: f'{row.created_date.year}-{"0" if row.created_date.month < 10 else ""}{row.created_date.month}'
col = dataframe.apply(fn, axis=1)
dataframe = dataframe.assign(month=col.values)
dataframe
```

```python
dataframe.groupby('transfer_type').count()
```

```python
filter_by_shipment_transaction_type = dataframe['shipment_transaction_type'] == 'Wholesale'
filtered_dataframe = dataframe[filter_by_shipment_transaction_type]
filter_by_transfer_type = dataframe['transfer_type'] == 'INCOMING_FROM_VENDOR'
filtered_dataframe = filtered_dataframe[filter_by_transfer_type]
len(filtered_dataframe)
```

```python
filtered_dataframe = filtered_dataframe.sort_values(by=['month'], ascending=False)
filtered_dataframe
```

```python
final_dataframe = filtered_dataframe[[
    'id',
    'created_date',
    'package_shipper_wholesale_price',
    'package_shipped_quantity',
    'month',
]]
final_dataframe
```

```python
fn = lambda row: row.package_shipper_wholesale_price * row.package_shipped_quantity if row.package_shipper_wholesale_price > 0.01 else 0
col = final_dataframe.apply(fn, axis=1)
final_dataframe = final_dataframe.assign(package_wholesale_gmv=col.values)
final_dataframe
```
